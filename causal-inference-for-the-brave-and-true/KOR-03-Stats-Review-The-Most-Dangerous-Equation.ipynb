# 03 - Stats Review: The Most Dangerous Equation

Howard Wainer는 2007년 유명한 [논문](http://assets.press.princeton.edu/chapters/s8863.pdf)을 통해 매우 위험한 방정식을 소개합니다.

"어떤 방정식은 알면 위험하고, 어떤 방정식은 모르면 위험하다. 어떤 방정식은 아는 것이 위험하다. 대표적으로 아인슈타인의 질량-에너지 등가 방정식 $E = MC^2$로, 이는 평범한 물질 안에 숨겨진 거대한 에너지 척도를 알려준다. \[...\] 나는 모를 때 위험한 방정식에 관심 있다. 이러한 방정식은 현상을 명확히 이해하게 해 주는데, 이들이 없으면 우리는 위험할 정도로 무지하게 된다."

Howard Wainer가 말하는 방정식은 드 무아브르(De Moivre) 방정식 또는 표준 오차에 대한 제곱근 규칙입니다.

$
SE = \dfrac{\sigma}{\sqrt{n}} 
$

$SE$는 표준 오차, $\sigma$는 표준 편차, $n$은 샘플 크기입니다. 용감하고 진실한 사람이 숙달해야 할 수학처럼 보입니다.

데이터를 통해 이 방정식을 모르는 것이 왜 매우 위험한지 살펴보겠습니다. 저는 3년 동안 다른 학교의 ENEM 점수(SAT와 유사한 브라질 표준화 고등학교 점수)에 대한 데이터를 수집했습니다. 데이터는 일부 처리되었으며 원본 데이터는 [Inep website](http://portal.inep.gov.br/web/guest/microdados#)에서 다운로드할 수 있습니다.

가장 성적이 좋은 학교는 눈에 띄는 특징이 있는데, 학생 수가 적다는 것입니다.

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")
```

```python
df = pd.read_csv("./data/enem_scores.csv")
df.sort_values(by="avg_score", ascending=False).head(10)
```

상위 1%의 학교의 학생수를 보면, 평균적인 학생 수가 적습니다.

```python
plot_data = (df
             .assign(top_school = df["avg_score"] >= np.quantile(df["avg_score"], .99))
             [["top_school", "number_of_students"]]
             .query(f"number_of_students<{np.quantile(df['number_of_students'], .98)}")) # remove outliers

plt.figure(figsize=(6,6))
sns.boxplot(x="top_school", y="number_of_students", data=plot_data)
plt.title("Number of Students of 1% Top Schools (Right)");
```

한 가지 자연스러운 결론은 학생 수가 적은 작은 학교일수록 학업 성취도가 높다는 것입니다. 선생님 한 명이 담당하는 학생 수가 적어지면 한 명 한 명에게 집중할 수 있어 이치에 맞아 보입니다. 하지만 이 결과가 드 무아브르 방정식과 무슨 관련이 있을까요? 그리고 왜 위험할까요?

이 결과로 중요한 결정을 내리면 위험해집니다. Howard는 말합니다:

"90년대에는 학교 규모를 줄이자는 주장이 인기를 끌었다. 작은 학교의 높은 성적은 과도하게 평가되어 많은 자선 단체와 정부 기관이 큰 학교를 분리하는 데 자금을 지원토록 만들었다."

사람들이 간과한 것은 하위 1%입니다. 확인해 보면 그들 또한 학생 수가 매우 적습니다!

```python
q_99 = np.quantile(df["avg_score"], .99)
q_01 = np.quantile(df["avg_score"], .01)

plot_data = (df
             .sample(10000)
             .assign(Group = lambda d: np.select([d["avg_score"] > q_99, d["avg_score"] < q_01],
                                                 ["Top", "Bottom"], "Middle")))
plt.figure(figsize=(10,5))
sns.scatterplot(y="avg_score", x="number_of_students", hue="Group", data=plot_data)
plt.title("ENEM Score by Number of Students in the School");
```

이 예시는 드 무아브르 방정식을 명확히 보여줍니다. 학생의 수가 많아질수록 (샘플이 많아질수록) 평균 점수는 더 정확해집니다. 학생 수가 적은 학교는 단지 우연으로 높은 점수나 낮은 점수를 받을 수 있습니다. 학생 수가 많아질수록 평균 점수는 실제와 가까워집니다. 드 무아브르 방정식은 실제 현상과 데이터 간 관계에 대한 근본적인 사실(항상 부정확함)을 알려줍니다. 문제는 얼마나 부정확한가?입니다.

통계학은 부정확성을 다루는 과학입니다. Taleb에 따르면 우리는 무작위성에 속기 쉽습니다.

> 확률은 주사위 확률과 같은 단순한 계산이 아닙니다. 확신의 부족을 받아들이고 무지를 다루는 방법을 개발하는 방법론입니다.

불확실성을 정량화하는 한 가지 방법은 **추정치의 분산**입니다. 분산은 관측치가 중앙값 혹은 가장 가능성이 높은 값에서 얼마나 멀어져 있는지를 나타냅니다. 드 무아브르 방정식에서처럼 불확실성은 우리가 관찰하는 데이터가 많아질수록 감소합니다. 많은 학생이 학교에서 우수한 성적을 낸다면, 정말 좋은 학교라 확신할 수 있습니다. 하지만 10명의 학생 중 8명이 좋은 성적을 내는 학교라면 의심할 필요가 있습니다. 우연히 학교에 평균 이상의 학생이 있을 수 있기 때문입니다.

위 그래프는 드 무아브르 방정식을 정확히 말해줍니다. 표본 크기가 작을수록 성적에 대한 추정치는 큰 차이를 보입니다. 표본 크기가 커지면 분산이 작아집니다. 우리가 인과 추론으로 추정하는 `ATE`를 포함한 모든 통계가 드 무아브르 방정식의 영향을 받습니다.

##  The Standard Error of Our Estimates

기초 통계학을 빠르게 정리해 보겠습니다. 분포, 분산, 표준 오차에 익숙하지 않으신 분은 계속 읽어보시고 충분히 이해되지 않으면 추가 자료를 찾아보세요. 통계학 입문에 관한 MIT 과정을 추천해 드립니다.

2장에서 평균 처치 효과 $E[Y_1 - Y_0]$는 처치된 것과 아닌 것 차이의 평균 $E[Y|T=1] - E[Y|T=0]$을 추정합니다. 우리는 온라인 수업에 대한 $ATE$를 계산했습니다. 온라인 수업은 성적에 악영향을 주는 결론을 얻었는데, 대면 수업보다 평균적으로 성적을 약 5점 떨어뜨립니다. 이제 이 5점이 통계적으로 얼마나 유의한지 확인해 봅시다.

먼저 $SE$를 계산합니다. 이미 우리는 샘플 크기인 $n$을 알고 있습니다. 표준 편차는 다음과 같이,

$$
\hat{\sigma}=\sqrt{\frac{1}{N-1}\sum_{i=1}^N (x_i-\bar{x})^2}
$$

$\bar{x}$는 $x$의 평균입니다. 다행히도 대부분 프로그램은 표준 편차를 이미 구현하고 있습니다. `pandas`의 [std](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.std.html)를 사용할 수 있습니다.

```python
data = pd.read_csv("./data/online_classroom.csv")
online = data.query("format_ol==1")["falsexam"]
face_to_face = data.query("format_ol==0 & format_blended==0")["falsexam"]

def se(y: pd.Series):
    return y.std() / np.sqrt(len(y))

print("SE for Online:", se(online))
print("SE for Face to Face:", se(face_to_face))
```

## Confidence Intervals

표준 오차는 신뢰 척도입니다. 이를 확실히 이해하기 위해서는 격동적이고 논쟁적인 통계 바다에 빠져봐야 합니다. 빈도주의 관점에서 데이터는 정확한 데이터 생성 프로세스의 표현일 뿐이라 주장합니다. 이는 추상적이고 이상적이며 변하지 않고 알려지지 않은 진정한 매개변수로 결정됩니다. 학생의 성적을 여러 번 수집할 수 있다면 개별 데이터 세트는 비슷하지만, 완전히 같지는 않을 것입니다. 마치 플라톤의 주장과 같습니다.

> 본질적으로 **하나이지만**, 여러 행위와 물질의 결합이나 조합으로 **여럿처럼 보인다** (주:플라톤 국가 5권 p.475)

이해를 위해 학생들의 시험 점수에 대한 진정한 분포를 알고 있다고 해보겠습니다. 진정한 분포는 평균이 74이고 표준 편차가 2인 정규 분포입니다. 이 정규 분포로 실험을 10,000번 수행하는데, 개별 실험은 500개의 샘플로 구성됩니다. 히스토그램으로 분포를 확인하면 개별 실험(500개 데이터로 이루어짐)의 평균이 실제 평균값인 74를 중심으로 분포되어 있음을 알 수 있습니다. 일부 실험의 평균은 74보다 낮고 일부는 더 높습니다.

```python
true_std = 2
true_mean = 74

n = 500
def run_experiment(): 
    return np.random.normal(true_mean,true_std, 500)

np.random.seed(42)

plt.figure(figsize=(8,5))
freq, bins, img = plt.hist([run_experiment().mean() for _ in range(10000)], bins=40, label="Experiment Means")
plt.vlines(true_mean, ymin=0, ymax=freq.max(), linestyles="dashed", label="True Mean", color="orange")
plt.legend();
```

우리는 지금 평균들의 평균에 관해 이야기하고 있습니다. 평균의 평균은 우연히 실제 평균보다 약간 낮거나 높을 수 있습니다. 즉, 실험의 평균이 플라토닉한 이상적인 평균과 일치한다고 확신할 수 없습니다. 하지만 **표준 오차를 활용해 95%의 실제 평균을 포함하는 구간은 만들 수 있습니다.**

실제로는 여러 데이터 세트로 같은 실험을 할수는 없으며, 우리는 보통 하나만 가지고 있습니다. 그래도 **신뢰 구간**이라 부르는 것은 만들 수 있습니다. 신뢰 구간에는 확률이 붙습니다. 95%가 일반적입니다. 확률은 실제 평균을 포함하는 서로 다른 연구의 가상 신뢰 구간을 알려줍니다. 예를 들어 유사한 연구에서 계산된 95% 신뢰 구간은 95%  확률로 실제 평균(true mean, platonic, ideal mean)을 포함합니다.

신뢰 구간을 계산하기 위해 **중심극한정리**를 사용합니다. 중심극한정리는 **실험의 평균이 결국 정규 분포를 따른다는 것을 말해줍니다.** 통계 이론으로 95%는 정규분포의 `2` 표준 편차 사이에 해당함을 알 수 있습니다. 엄밀히 따지면 $1.96$이지만 $2$도 괜찮습니다.

![normal_density](./data/img/stats-review/normal_dist.jpeg)

평균의 표준 오차는 평균의 분포에 대한 추정치를 제공합니다. 실험 내 평균에서 표준 오차에 2를 곱한 값을 빼거나 더하면 실제 평균에 대한 95% 신뢰 구간을 구성할 수 있습니다.

```python
np.random.seed(321)
exp_data = run_experiment()
exp_se = exp_data.std() / np.sqrt(len(exp_data))
exp_mu = exp_data.mean()
ci = (exp_mu - 2 * exp_se, exp_mu + 2 * exp_se)
print(ci)
```

```python
x = np.linspace(exp_mu - 4*exp_se, exp_mu + 4*exp_se, 100)
y = stats.norm.pdf(x, exp_mu, exp_se)
plt.plot(x, y)
plt.vlines(ci[1], ymin=0, ymax=1)
plt.vlines(ci[0], ymin=0, ymax=1, label="95% CI")
plt.legend()
plt.show()
```

물론 신뢰 구간을 95%로 해야 할 필요는 없습니다. 99%에 해당하는 표준 편차를 찾아 99% 신뢰 구간을 만들 수도 있습니다.

`python`의 `ppf` 함수는 CDF의 역을 제공합니다. 95% CI를 찾았을 때처럼 표준 오차에 $2$를 곱하는 대신, `z`를 곱하면 99% CI가 됩니다. `ppf(0.5)`는 $0.0$을 반환하는데, 정규분포의 50%가 $0.0$보다 작기 때문입니다. 같은 이유로, 99.5%에 대응하는 `z`를 얻게 된다면 분포의 99.5%는 `z` 아래로 떨어집니다. 즉, 전체 중 0.5%가 `z`보다 큽니다.

```python
from scipy import stats
z = stats.norm.ppf(.995)
print(z)
ci = (exp_mu - z * exp_se, exp_mu + z * exp_se)
ci
```

```python
x = np.linspace(exp_mu - 4*exp_se, exp_mu + 4*exp_se, 100)
y = stats.norm.pdf(x, exp_mu, exp_se)
plt.plot(x, y)
plt.vlines(ci[1], ymin=0, ymax=1)
plt.vlines(ci[0], ymin=0, ymax=1, label="99% CI")
plt.legend()
plt.show()
```

이를 활용하여 온라인 수업과 대면 수업 그룹 모두의 평균 점수에 대한 신뢰 구간을 계산할 수 있습니다.

```python
def ci(y: pd.Series):
    return (y.mean() - 2 * se(y), y.mean() + 2 * se(y))

print("95% CI for Online:", ci(online))
print("95% for Face to Face:", ci(face_to_face))
```

두 그룹 간 95% CI는 겹치지 않습니다. 대면 수업 그룹의 CI 최솟값은 온라인 수업 그룹의 CI 최댓값보다 큽니다. 따라서 분석 결과가 우연이 아니며 대면 수업이 온라인 수업보다 낫다는 증거입니다. 즉 대면 수업에서 온라인수업으로 전환할 때 학업성적에 상당한 인과적 감소가 발생합니다.

요약하면 신뢰 구간은 추정치 주변에 불확실성을 두는 방법입니다. 표본 크기가 작을수록 표준 오차는 크고 신뢰 구간이 넓습니다. 신뢰 구간은 계산하기 쉬우므로 누군가가 신뢰 구간을 보여주지 않는다면 나쁜 의도를 가지고 있거나, 지식이 부족하다는 신호입니다. (모두 우려되는 사항입니다.) 불확실성 측정이 없는 값은 항상 의심해야 합니다.

</figure>

신뢰 구간은 보기보다 해석이 어렵습니다. 예를 들어 특정 95% 신뢰 구간에 95% 모집단 평균이 포함되어 있다고 말하면 안 됩니다. 신뢰 구간을 사용하는 빈도 통계에서는 모집단 평균은 실제 모집단 상수로 간주합니다. 모집단 평균은 신뢰 구간에 있거나 없을 수 있습니다. 즉 특정 신뢰 구간이 진짜 평균을 포함하거나 포함하지 않습니다. 만약 포함한다고 말한다면 확률은 95%가 아니라 100%가 될 것입니다. 그렇지 않으면 확률이 0%가 됩니다. 대신 신뢰 구간 95%는 신뢰 구간이 진짜 평균을 포함하는 빈도를 나타냅니다. 95%는 특정 구간 자체가 아니라 95% CI를 계산하는 데 사용되는 알고리즘에 대한 신뢰입니다.

경제학자(통계학자들은 눈을 돌려주세요.)로서 저는 이러한 순수주의가 유용하지 않다고 생각합니다. 실제로 신뢰 구간에 95%의 실제 평균이 포함되어 있다고 말하는 사람을 볼 수 있습니다. 틀리긴 하지만 여전히 추정치에 불확실성을 말하고 있어 그다지 해롭지는 않습니다. 또한, 베이지안 통계와 같이 신뢰 구간 대신 가능 구간을 사용해 해당 구간이 95%의 분포 평균을 포함한다고 할 수 있습니다. 제가 본 바로는 적절한 표본 크기를 가진 베이지안 확률 구간은 신뢰 구간과 유사하며, 빈도론자들 또한 인정합니다. 제 의견이 타당하다면 당신의 신뢰 구간에 대해 자유롭게 말하세요. 신뢰 구간이 95%의 진짜 평균을 포함한다고 말해도 상관없습니다. 다만 당신의 추정치 주변에 놓는 것을 잊지 마세요. 그렇지 않으면 당신은 바보처럼 보일 것입니다.

## Hypothesis Testing

불확실성을 통합하는 또 다른 방법은 가설 검정을 하는 것입니다. 평균의 차이가 통계적으로 0(또는 다른 값)과 다른가요? 두 독립 정규 분포의 합 또는 차도 정규 분포입니다. 결과의 평균은 두 분포 사이의 합 또는 차이며 전체 분산은 항상 각 분산의 합이 됩니다.

$
N(\mu_1, \sigma_1^2) - N(\mu_2, \sigma_2^2) = N(\mu_1 - \mu_2, \sigma_1^2 + \sigma_2^2)
$

$
N(\mu_1, \sigma_1^2) + N(\mu_2, \sigma_2^2) = N(\mu_1 + \mu_2, \sigma_1^2 + \sigma_2^2)
$

```python
np.random.seed(123)
n1 = np.random.normal(4, 3, 30000)
n2 = np.random.normal(1, 4, 30000)
n_diff = n2 - n1
sns.distplot(n1, hist=False, label="$N(4,3^2)$")
sns.distplot(n2, hist=False, label="$N(1,4^2)$")
sns.distplot(n_diff, hist=False, label=f"$N(1,4^2) - (4,3^2) = N(-3, 5^2)$")
plt.legend()
plt.show()
```

두 그룹의 평균 분포를 취하고 다른 그룹에서 하나를 빼면 세 번째 분포를 갖습니다. 이 최종 분포의 평균은 평균의 차이가 되고, 이 분포의 표준 편차는 표준 편차의 합의 제곱근이 됩니다.

$
\mu_{diff} = \mu_1 - \mu_2
$

$
SE_{diff} = \sqrt{SE_1 + SE_2} = \sqrt{\sigma_1^2/n_1 + \sigma_2^2/n_2}
$

온라인 수업 예시로 돌아가겠습니다. 우리는 두 그룹 간 차이의 분포를 만듭니다. 물론 95% CI를 계산하는 것은 간단합니다.

```python
diff_mu = online.mean() - face_to_face.mean()
diff_se = np.sqrt(face_to_face.var()/len(face_to_face) + online.var()/len(online))
ci = (diff_mu - 1.96*diff_se, diff_mu + 1.96*diff_se)
print(ci)
```

```python
x = np.linspace(diff_mu - 4*diff_se, diff_mu + 4*diff_se, 100)
y = stats.norm.pdf(x, diff_mu, diff_se)
plt.plot(x, y)
plt.vlines(ci[1], ymin=0, ymax=.05)
plt.vlines(ci[0], ymin=0, ymax=.05, label="95% CI")
plt.legend()
plt.show()
```

온라인 수업 그룹과 대면 수업 그룹 간의 실제 차이는 $-8.37$과 $-1.44$ 사이에 있다고 95% 신뢰할 수 있습니다. **z**값을 얻기 위해는 차이의 평균을 차이의 $SE$로 나눌 수 있습니다.

$
z = \dfrac{\mu_{diff} - H_{0}}{SE_{diff}} = \dfrac{(\mu_1 - \mu_2) - H_{0}}{\sqrt{\sigma_1^2/n_1 + \sigma_2^2/n_2}}
$

$H_0$는 테스트하고자 값입니다.

`z` 통계량은 관측된 차이가 얼마나 극단적인지를 나타내는 척도입니다. 평균의 차이가 $0$과 통계적으로 다르다는 가설을 검정하기 위해 모순되는 가정을 사용할 것입니다. 먼저 온라인 수업과 대면 수업 간 차이가 없다고 가정하겠습니다. 이를 `null hypothesis`(귀무가설) 혹은 $H_0$라 합니다. "만약 정말로 차이가 없다면, 이를 관측할 가능성이 있는가?" 를 물어봅니다. 귀무가설로 문제를 `z` 통계량이 $0$에서 얼마나 멀리 떨어져 있는지 확인하는 문제로 바꿀 수 있습니다.

$H_0$의 `z` 통계량은 표준 정규 분포를 따릅니다. 따라서 차이가 실제로 0이면 평균 95%인 2 표준 편차 내에서 `z` 통계량을 볼 수 있습니다. 실제로는 `z`가 2 표준 편차를 넘어가므로 95% 신뢰도로 귀무가설을 기각할 수 있습니다.

실제 예제에서 되는지 확인해 보겠습니다.

```python
z = diff_mu / diff_se
print(z)
```

```python
x = np.linspace(-4,4,100)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, label="Standard Normal")
plt.vlines(z, ymin=0, ymax=.05, label="Z statistic", color="C1")
plt.legend()
plt.show()
```

`z`는 꽤 극단적으로 보입니다. `z`가 2보다 크면, 극단적인 값을 볼 확률이 5% 미만임을 의미합니다. 따라서 온라인 수업은 통계적으로 상당한 학업 성취도의 감소를 일으킨다는 결론을 얻습니다.

가설 검정의 흥미로운 점은 처치된 그룹과 그렇지 않은 그룹의 95% CI가 겹치는지 확인하는 것보다 덜 보수적이라는 것입니다. 즉, 두 그룹의 신뢰 구간이 겹치는 경우도 통계적으로 유의할 수 있습니다. 예를 들어 대면 수업 그룹의 평균 점수는 80점, 표준 오차가 4이고, 온라인 수업 그룹의 평균 점수는 71점, 표준 오차가 2라고 해 보겠습니다.

```python
cont_mu, cont_se = (71, 2)
test_mu, test_se = (80, 4)

diff_mu = test_mu - cont_mu
diff_se = np.sqrt(cont_se**2 + test_se**2)

print("Control 95% CI:", (cont_mu-1.96*cont_se, cont_mu+1.96*cont_se))
print("Test 95% CI:", (test_mu-1.96*test_se, test_mu+1.96*test_se))
print("Diff 95% CI:", (diff_mu-1.96*diff_se, diff_mu+1.96*diff_se))
```

각 그룹에 대한 신뢰 구간은 서로 겹칩니다. 온라인 그룹의 95% CI 상한은 74.92이고 대면 그룹의 하한은 72.16입니다. 그래도 그룹 간 차이에 대한 95% 신뢰 구간에는 $0$이 포함되지 않습니다. 개별 신뢰 구간이 겹치더라도 그 차이는 여전히 통계적으로 $0$과 다를 수 있습니다.

## P-values

앞에서 온라인 수업 그룹과 대면 수업 그룹의 차이가 실제로 $0$이라면 $0$을 관찰할 확률은 5% 미만이라고 말했습니다. 극단적인 값을 관찰할 가능성이 얼마인지 정확하게 알 수 있을까요? `p-value`을 확인하면 됩니다!

신뢰 구간(실제로 가장 빈번한 통계치입니다.)과 마찬가지로 `p-value`의 정의는 어려울 수 있습니다. 위키피디아에서는 다음과 같이 정의됩니다. "`p-value`는 귀무가설이 맞다고 가정할 때 얻은 결과보다 극단적인 결과가 실제로 관측될 확률이다."

간단하게 말하자면 귀무가설이 참일 때, 극단적인 값을 데이터를 볼 확률이 `p-value`입니다. 귀무가설이 참이면 측정값을 볼 가능성이 얼마나 낮은지 측정합니다. 하지만 자연스럽게 귀무가설이 참일 확률과 혼동되는 경우가 많습니다.  `p-value`는 $P(H_0|data)$가 아니라 $P(data|H_0)$입니다.

이론적으로는 복잡하지만 실제로는 간단히 계산할 수 있습니다.

![p_value](./data/img/stats-review/p_value.png)

`p-value`를 얻으려면 `z` 통계량 앞이나 뒤의 표준 정규 분포 면적을 계산해야 합니다. 다행히도 우리는 계산을 해줄 컴퓨터가 있습니다. 표준 정규 분포의 CDF에 z 통계량을 넣어 확인합니다.

```python
print("P-value:", stats.norm.cdf(z))
```

95% 또는 99%와 같은 신뢰 수준을 지정할 필요가 없다는 점에서 `p-value`는 흥미롭습니다. `p-value`를 계산하면 검정이 통과하거나 실패할 신뢰도를 정확히 알게 됩니다. 예를 들어 p-값이 0.0027이면 0.2% 수준까지 유의하다는 의미입니다. 따라서 차이의 95% CI와 99% CI에는 0이 포함되지 않지만 99.9% CI에는 0이 포함됩니다. 차이가 0인 경우, 극한의 `z` 통계량을 관측할 확률은 0.2%에 불과합니다.

```python
diff_mu = online.mean() - face_to_face.mean()
diff_se = np.sqrt(face_to_face.var()/len(face_to_face) + online.var()/len(online))
print("95% CI:", (diff_mu - stats.norm.ppf(.975)*diff_se, diff_mu + stats.norm.ppf(.975)*diff_se))
print("99% CI:", (diff_mu - stats.norm.ppf(.995)*diff_se, diff_mu + stats.norm.ppf(.995)*diff_se))
print("99.9% CI:", (diff_mu - stats.norm.ppf(.9995)*diff_se, diff_mu + stats.norm.ppf(.9995)*diff_se))
```

## Key Ideas

드 무아브르 방정식을 아는 것이 얼마나 중요한지를 살펴보았고, 추정치 주변에 불확실성을 두는 방법을 배웠습니다. 온라인 수업 예시를 통해 분석 결과가 통계적으로 유의하다는 사실을 확인했습니다. 평균의 신뢰 구간을 비교하고, 차이에 대한 신뢰 구간을 살펴보고, 가설 검정을 수행하고, `p-value`를 살펴봤습니다. A/B 테스트를 수행하는 함수로 이 장을 마무리하겠습니다.

```python
def AB_test(test: pd.Series, control: pd.Series, confidence=0.95, h0=0):
    mu1, mu2 = test.mean(), control.mean()
    se1, se2 = test.std() / np.sqrt(len(test)), control.std() / np.sqrt(len(control))
    
    diff = mu1 - mu2
    se_diff = np.sqrt(test.var()/len(test) + control.var()/len(control))
    
    z_stats = (diff-h0)/se_diff
    p_value = stats.norm.cdf(z_stats)
    
    def critial(se): return -se*stats.norm.ppf((1 - confidence)/2)
    
    print(f"Test {confidence*100}% CI: {mu1} +- {critial(se1)}")
    print(f"Control {confidence*100}% CI: {mu2} +- {critial(se2)}")
    print(f"Test-Control {confidence*100}% CI: {diff} +- {critial(se_diff)}")
    print(f"Z Statistic {z_stats}")
    print(f"P-Value {p_value}")
        
AB_test(online, face_to_face)
```

함수 `AB_test`는 충분히 일반적이므로 다른 귀무가설도 검정할 수 있습니다. 예를 들어 온라인 수업과 대면 수업의 차이가 `-1`이라는 것을 기각할 수 있나요? 결과를 보면 95%의 확신을 가지고 그 차이가 `-1`보다 더 중요하다고 말할 수 있습니다. 하지만 99%의 확신을 가지고 말할 수는 없습니다.

```python
AB_test(online, face_to_face, h0=-1)
```
