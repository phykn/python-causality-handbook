## Answering a Different Kind of Question

현재 `Machine Learning`(기계학습)은 예측 문제를 잘 해결합니다. Ajay Agrawal, Joshua Gans, Avi Goldfarb의 책 "Prediction Machines"에서 언급한 것처럼 "인공지능의 새로운 물결은 지능 자체 보다는 지능의 구성요소를 제공합니다." `Machine Learning`으로 아름다운 작업을 할 수 있습니다. 우리가 해야 할 일은 문제를 예측문제로 바꾸는 것뿐 입니다. 영어를 포르투갈어로 번역하고 싶나요? 영어 문장을 넣고 포르투갈어 문장을 예측하는 ML 모델을 만듭니다. 얼굴인식 프로그램을 만들고 싶나요? 사진에서 얼굴 위치를 예측하는 ML 모델을 만듭니다. 자율주행 자동차를 만들고 싶나요? ML 모델로 차량 주변의 이미지와 센서값을 받아 휠의 방향, 브레이크, 가속기의 압력을 예측하세요.

하지만 ML은 만병통치약이 아닙니다. 학습된 데이터 경계 안에서만 잘 작동하며 여기서 약간만 벗어나더라도 실패합니다. Prediction Machines의 다른 예를 들자면, "많은 산업에서 낮은 가격은 낮은 판매와 관련됩니다. 호텔 가격은 비수기에 낮지만, 성수기에 높아집니다. 따라서 가격을 올리면 더 많은 방이 팔릴 것이라는 단순한 예측이 가능합니다."

ML은 역인과 문제를 해결하지 못하기로 악명 높습니다. 인과문제는 경제학자들이 `counterfactual`(반사실)이라 부르는 "만약" 이라는 질문에 대답하는 것입니다. 만약 상품가격을 바꾼다면 어떨까요? 만약 저지방 다이어트 대신 저설탕 다이어트를 하면 어떨까요? 은행에서 일한다면 고객의 신용한도를 바꾸는 것이 수익을 낼 수 있을지 알아야 합니다. 정부에서 일한다면 교육 시스템을 더 좋게 만드는 방법이 필요합니다. 모든 학생에게 태블릿을 제공해야 할까요? 아니면 구식 도서관을 지어야 할까요?

이러한 질문의 핵심은 인과성 입니다. 매출을 올리는 방법과 같은 일상적인 문제에는 인과관계에 대한 질문이 스며들어 있습니다. 인생에서 성공하기 위해 비싼 학교에 가야 할까요? (학위가 연봉을 올려줄까요?) 이민은 직업을 가질 가능성을 낮추나요? (이민은 실업률을 높일까요?) 가난한 사람들을 지원하면 범죄율이 낮아지나요? 분야와 관계없이 이러한 문제는 산재해 있습니다. 우리는 인과 문제에 답해야 합니다. 상관관계에 기반을 둔 ML 예측으로는 해결할 수 없습니다.

인과문제에 대한 대답은 생각보다 어렵습니다. 아마도 여러분은 "상관은 인과가 아니다." 라는 말을 알고 있을 겁니다. 사실 `association`(상관)이 `causation`(인과)가 아닌지 설명하는 것이 더 중요합니다. 도입은 이것으로 마칩니다. 책의 나머지 부분에 대해서는 **상관관계를 인과관계로 만드는 방법**을 찾는 데 전념할 것입니다.

## When Association IS Causation

직관적으로는 왜 상관관계가 인과관계가 아닌지 알고 있습니다. 만약 태블릿을 받는 학교가 그렇지 않은 학교보다 평균 성적이 높다면, 태블릿을 제공하는 학교들이 더 부유했기 때문이라고 지적할 수 있습니다. 태블릿과 관계없이 성적이 더 좋을 수도 있습니다. 따라서 태블릿 제공이 성적을 올린다고 할 수는 없습니다. 그저 태블릿과 높은 성적이 관련 있다고 할 수 있을 뿐입니다.

````python
import pandas as pd
import numpy as np
from scipy.special import expit
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import style

style.use("fivethirtyeight")

np.random.seed(123)
n = 100
tuition = np.random.normal(1000, 300, n).round()
tablet = np.random.binomial(1, expit((tuition - tuition.mean()) / tuition.std())).astype(bool)
enem_score = np.random.normal(200 - 50 * tablet + 0.7 * tuition, 200)
enem_score = (enem_score - enem_score.min()) / enem_score.max()
enem_score *= 1000

data = pd.DataFrame(dict(enem_score=enem_score, Tuition=tuition, Tablet=tablet))
````

````python
plt.figure(figsize=(6,8))
sns.boxplot(y="enem_score", x="Tablet", data=data).set_title('ENEM score by Tablet in Class')
plt.show()
````

구체적으로 표현하기 위해 몇 가지 `notation`을 만들겠습니다. 앞으로 우리의 일상적인 언어가 될 것입니다. 다른 용감하고 진정한 "인과 전사"를 식별하는 데 사용할 공용어라 생각하세요. 다가올 많은 전투에서 우리의 함성이 될 것입니다.

$T_i$는 `unit`(단위 데이터, table의 경우 1개 row에 해당함) $i$에 대한 `treatment`(처치)입니다.

$
T_i=\begin{cases}
1 \ \text{if unit i received the treatment}\\
0 \ \text{otherwise}\\
\end{cases}
$

`treatment`이 약이나 의료 분야에만 쓰이는 것은 아닙니다. 단지 개입을 나타내기 위해 사용하는 용어일 뿐입니다. 태블릿 예시에서 `treatment`는 태블릿 지급 여부입니다. 참고로 `treatment`는 $D$로 표기하기도 합니다.

$Y_i$는 `unit` $i$에 대한 관측 결과입니다.

`outcome`은 관심 있는 결과 값입니다. 우리는 `treatment`가 `outcome`에 어떤 영향을 주는지 알고 싶습니다. 태블릿 예시에서 `outcome`은 성적입니다.

이제 재미있는 부분입니다. **인과추론의 근본적인 어려움**은 같은 `unit`에서 모든 `treatment` 결과를 알 수 없는 점입니다. 갈라진 두 개의 길 중 선택한 하나의 길에 무엇이 있는지 알 수 있을 뿐입니다. 로버트 프로스트의 시처럼요.
```
노란 숲 속에 길이 두 갈래로 났었습니다.  
나는 두 길을 다 가지 못하는 것을 안타깝게 생각하면서,  
오랫동안 서서 한 길이 굽어 꺾여 내려간 데까지,  
바라다볼 수 있는 데까지 멀리 바라다보았습니다.  
```

이 문제를 다루기 위해 `potential outcome`(잠재 결과)에 대해 많은 이야기를 나눌 것입니다. 실제로 일어나지 않은 일은 잠재해 있습니다. `potential outcome`은 만약 `treatment`가 적용되었다면 **어떻게 되었을까**를 나타냅니다. 발생한 `potential outcome`은 `factual`(사실), 발생하지 않은 `outcome`은 `counterfactual`(반사실)이라 부릅니다.

표기는 아래첨자를 사용합니다.

$Y_{0i}$는 `treatment`가 없는 `unit` $i$의 `potential outcome`입니다.

$Y_{1i}$는 `treatment`가 있는 **동일한** `unit` $i$의 `potential outcome`입니다.

`potential outcome`은 $Y_i(t)$로 표기하기도 합니다. $Y_{0i} = Y_i(0)$, $Y_{1i} = Y_i(1)$ 입니다. 이 책에서는 아래첨자 표기법을 사용합니다.

![img](./data/img/intro/potential_outcomes.png)

$Y_{1i}$는 태블릿을 지급한 학교 학생 $i$의 성적입니다. 사실이든 반사실이든 $Y_{1i}$은 같습니다. 학생 $i$가 태블릿을 받았다면 $Y_{1i}$를 관찰할 수 있습니다. 반대의 경우 $Y_{0i}$를 관찰할 수 있습니다. $Y_{0i}$을 관찰했다면 $Y_{1i}$는 알 수 없습니다. 이 경우 $Y_{1i}$는 `counterfactual`(반사실적) `potential outcome`입니다.

`potential outcome`으로 `individual treatment effect`(개별 처치 효과)를 정의할 수 있습니다.

 $Y_{1i} - Y_{0i}$
 
물론 근본적으로  `potential outcome` 중 하나만 알 수 있기에 개별 처치 효과를 알 수는 없습니다. 당분간 개별 처치 효과보다 더 쉬운 것에 초점을 맞추겠습니다. 아래와 같이 정의된 **average treatment effect**(평균 처치 효과)를 살펴봅시다.

$ATE = E[Y_1 - Y_0]$

`E[...]`는 기댓값입니다. 추정하기 쉬운 또 다른 양은 **average treatment effect on the treated**(처치된 평균 처치 효과)입니다:

$ATT = E[Y_1 - Y_0 | T=1]$

현실적으로 모든 `potential outcome`은 볼 수 없지만, 가능하다고 생각해 봅시다. 인과추론의 신이 우리를 가엾게 여겨 모든 `potential outcome`을 알려주었습니다. 4개 학교에 대한 데이터를 수집해 보겠습니다. 태블릿 지급 여부에 따른 성적을 모두 알고 있습니다. 태블릿 지급 여부가 `treatment`이므로 학생에게 태블릿을 제공했다면 $T=1$입니다. $Y$는 성적입니다.

````python
pd.DataFrame(dict(
    i= [1,2,3,4],
    y0=[500,600,800,700],
    y1=[450,600,600,750],
    t= [0,0,1,1],
    y= [500,600,600,750],
    te=[-50,0,-200,50],
))
````

$ATE$는 마지막 `column`의 평균, 즉 `treatment effect`의 평균입니다.

$ATE=(-50 + 0 - 200 + 50)/4 = -50$

이는 태블릿 지급이 평균적으로 성적을 50점 낮춘다는 것을 의미합니다. $ATT$는 $T=1$인 마지막 `column`의 평균입니다.

$ATT=(- 200 + 50)/2 = -75$

`treatment`가 적용된 학교는 태블릿이 성적을 평균적으로 75점 감소시킴을 의미합니다. 실제로는 직접 계산할 수 없습니다. 현실적인 데이터는 다음과 같습니다.

````python
pd.DataFrame(dict(
    i= [1,2,3,4],
    y0=[500,600,np.nan,np.nan],
    y1=[np.nan,np.nan,600,750],
    t= [0,0,1,1],
    y= [500,600,600,750],
    te=[np.nan,np.nan,np.nan,np.nan],
))
````

이상적인 방법은 아니지만 처치된 데이터의 평균과 처치 받지 않은 데이터의 평균과 비교할 수는 없을까요? $ATE = (600+750)/2 - (500 + 600)/2 = 125$가 되면 안 되나요? 절대 안 됩니다! 결과가 얼마나 다른지 보세요. 당신은 상관관계를 인과관계로 판단한 가장 큰 죄를 지었습니다. 죄를 이해하기 위해 인과추론의 악당을 소개합니다.

## Bias

편향은 상관관계가 인과관계가 아니게 만듭니다. 직관적으로 이해할 수 있습니다. 태블릿 예시를 다시 정리해 보겠습니다. 누가 태블릿을 지급하는 학교의 성적이 더 높다고 주장한다면, 태블릿과 무관하데 더 높은 성적을 얻을 수 있을 것이라 반박할 수 있습니다. 아마도 원인은 학교의 재정상태일 것입니다. 학교는 더 나은 선생님들에게 많은 돈을 지급하고, 더 좋은 교실을 제공할 수 있습니다. 따라서 처치 받은 (태블릿을 지급한) 학교는 그렇지 않은 학교와 비교할 수 없습니다.

`potential outcome` 표기는 처치된 $Y_0$이 그렇지 않은 $Y_0$과 다르다는 것을 보여줍니다. 처치된 $Y_0$는 **반사실적**입니다. 관찰할 수는 없지만 추론은 가능합니다. 태블릿이 지급된 학교의 $Y_0$은 그렇지 않은 학교의 $Y_0$보다 더 클 것이라 할 수 있습니다. 태블릿을 지급할 여유가 있는 학교는 성적을 위해 다른 요소(더 좋은 선생님, 더 좋은 교실)를 제공할 수 있기 때문입니다. 잠시 쉬어갑시다. `potential outcome`을 이해하는 데는 시간이 좀 걸립니다.

상관관계가 왜 인과관계가 아닌지 기초 수학으로 확인해 봅시다. 상관관계는 $E[Y|T=1]-E[Y|T=0]$로 측정합니다. 태블릿 예제에서는 태블릿을 지급한 학교의 평균 점수에서 태블릿을 지급하지 않은 학교의 평균 점수를 뺀 값입니다. 반면 인과관계는 $E[Y_1 - Y_0]$로 측정합니다.

상관성을 측정하고 관찰된 `outcome`을 `potential outcome`으로 대체해 보겠습니다. 처치된 경우 `outcome`은 $Y_1$입니다. 그렇지 않으면 $Y_0$입니다.

$
E[Y|T=1] - E[Y|T=0] = E[Y_1|T=1] - E[Y_0|T=0]
$

이제 $E[Y_0|T=1]$를 더하고 빼겠습니다. $E[Y_0|T=1]$은 `counterfactual outcome`입니다. `treatment`가 없을 때 만약 `treatment`를 적용한다면 결과가 어땠을지 말해줍니다.

$
E[Y|T=1] - E[Y|T=0] = E[Y_1|T=1] - E[Y_0|T=0] + E[Y_0|T=1] - E[Y_0|T=1]
$

수식을 정렬하고 몇 가지 기댓값을 합칩니다.

$
E[Y|T=1] - E[Y|T=0] = \underbrace{E[Y_1 - Y_0|T=1]}_{ATT} + \underbrace{\{ E[Y_0|T=1] - E[Y_0|T=0] \}}_{BIAS}
$

이 간단한 수식은 인과질문에서 마주치는 모든 문제를 다룹니다. 꼭 이해하고 넘어가라고 당부드리고 싶습니다. 만약 문신을 한다면 좋은 후보가 될 것입니다. 성경과 같이 깊이 간직하고 이해해야 합니다. 더 깊이 살펴봅시다. 의미를 몇 가지로 나누어 보겠습니다. 첫째로 방정식은 상관관계가 인과관계가 아닌 이유를 알려줍니다. 상관관계는 처치 항에 편향을 더한 것과 같습니다. **편향은 `treatment`가 적용되기 이전 실험군과 대조군의 차이로 주어집니다.** 이제 태블릿이 성적을 높인다는 주장에 대해 의심스러운 점을 정확히 말할 수 있습니다. 태블릿 예시에서는 $E[Y_0|T=0] < E[Y_0|T=1]$입니다. 즉, 태블릿을 지급할 여유 있는 학교는 그렇지 못한 학교보다 **태블릿 지급 여부와 관계없이** 성적이 높습니다.

왜 편향이 생길까요? `confounding`(교란자)에 대해서는 나중에 이야기하겠지만, 지금은 통제할 수 없는 많은 것들이 `treatment`와 함께 바뀌어 편향이 생긴다고 이해할 수 있습니다. 실험군과 대조군은 `treatment`에만 차이가 있는 것이 아닙니다. 수업료, 장소, 선생님도 다릅니다. 태블릿이 성적을 올려준다고 말하려면 비슷한 학교끼리 비교해야 합니다.

````python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Tuition", y="enem_score", hue="Tablet", data=data, s=70).set_title('ENEM score by Tuition Cost')
plt.show()
````

정답을 알아봅시다. 이제 상관관계를 인과관계로 만들기 위해 무엇이 필요한지 압니다. **$E[Y_0|T=0] = E[Y_0|T=1]$이라면 상관관계는 인과관계입니다!** 이 방정식은 강력히 주장합니다. $E[Y_0|T=0] = E[Y_0|T=1]$는 `treatment` 전후에 실험군과 대조군이 유사하다는 의미입니다. 만약 처치되지 않았을 때 $Y_0$을 관찰할 수 있다면 두 경우 결과가 같을 것입니다. 수학적으로는 편향항이 사라집니다.

$
E[Y|T=1] - E[Y|T=0] = E[Y_1 - Y_0|T=1] = ATT
$

또한, 처치된 것과 아닌 것이 `treatment`만 차이난다면 $E[Y_0|T=0] = E[Y_0|T=1]$이며 인과 영향 정도는 (매우 유사하므로) `treatment` 여부에 상관없이 같습니다.

$
\begin{align}
E[Y_1 - Y_0|T=1] &= E[Y_1|T=1] - E[Y_0|T=1] \\
&= E[Y_1|T=1] - E[Y_0|T=0] \\
&= E[Y|T=1] - E[Y|T=0]
\end{align}
$

이 경우 **평균의 차이가 인과관계입니다.**

$
E[Y|T=1] - E[Y|T=0] = ATT
$

또한, `treatment` 그룹과 아닌 그룹이 `treatment` 여부만 다르다면 $E[Y_1|T=0] = E[Y_1|T=1]$를 얻습니다. 즉, 실험군과 대조군이 `treatment`에 대한 반응이 유사한지 확인할 수 있습니다. 이제 **`treatment` 이후 교환**도 가능합니다. 이 경우 $E[Y_1 - Y_0|T=1]=E[Y_1 - Y_0|T=0]$이며,
 
$
E[Y|T=1] - E[Y|T=0] = ATT = ATE
$

너무나 중요해서 아래 그림과 함께 다시 한번 살펴보겠습니다. 처치된 그룹과 아닌 그룹을 단순 평균하여 비교하면 다음과 같습니다. (파란색 점은 처치 받지 못했습니다.)

![img](./data/img/intro/anatomy1.png)

두 그룹 간의 `outcome` 차이의 원인은 두 가지입니다.

1. `treatment effect`입니다. 성적 향상은 태블릿 제공이 원인입니다.
2. 성적 차이는 등록금 때문일 수 있습니다. 두 그룹 차이는 교육비의 차이이며 태블릿 자체는 원인이 아닙니다.
   
`individual treatment effect`(개별 치리 효과)는 같은 `unit`에 대해 `treatment`에 따른 이론적인 `outcome` 차이입니다. 실제 `treatment effect`는 아래 왼쪽 그림과 같이 `potential outcome`을 관찰할 수 있는 신과 같은 능력이 있을 때만 얻을 수 있습니다. `counterfactual outcome`으로 밝은색으로 표시합니다.

![img](./data/img/intro/anatomy2.png)

오른쪽 그림은 편향을 보여줍니다. 모든 사람이 처치를 받지 않도록 하면 편향을 얻을 수 있습니다. 이 경우 $T_0$인 `potential outcome`만 남는데 처치된 그룹과 그렇지 않은 그룹의 차이가 편향입니다. 편향이 크다면 `treatment`가 아닌 요인이 `outcome`을 다르게 하는 원인입니다. 편향은 실제 `treatment effect`를 왜곡시킵니다.

편향이 없는 가상의 상황과 대조해 봅시다. 태블릿이 무작위로 지급된다고 가정해봅시다. 학교와 상관없이 태블릿을 받을 가능성은 같습니다. `treatment`는 학교 전반에 걸쳐 잘 분배될 것입니다.

![img](./data/img/intro/anatomy3.png)

이 경우 처치된 그룹과 아닌 그룹의 `outcome`차이가 평균적인 인과효과입니다. `treatment`외에 `outcome`을 다르게 하는 원인이 없기 때문입니다. 모든 차이는 `treatment`에 의한 것입니다. 편향이 존재하지 않습니다.

![img](./data/img/intro/anatomy4.png)

$Y_0$들만 관찰할 수 있게 (모든 사람이 처치 받지 않게) 설정한다면 두 그룹 간 차이를 찾을 수 없습니다.

편향은 인과추론을 대단히 어렵게 합니다. 편향을 제거하여 두 그룹을 비교하는 방법은 평균적인 `treatment effect`만을 알려줍니다. 궁극적으로 인과추론은 망상과 잘못된 해석을 제거하여 세상이 어떻게 돌아가는지 알아내는 것입니다. 이점을 이해했다면 편향을 없애는 강력한 방법을 터득하여 인과효과를 계산해 봅시다.
